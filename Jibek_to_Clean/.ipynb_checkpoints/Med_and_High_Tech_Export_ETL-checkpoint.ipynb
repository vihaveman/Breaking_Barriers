```python
import pandas as pd
import os
```

```python
mhtech_export= "Med_and_High_Tech_Export.csv"
mhtech_export_df = pd.read_csv(mhtech_export, sep="\t")
mhtech_export_df.head(5)
```

```python
list(mhtech_export_df)
```

```python
# Get the contents of row 2
row_2_data = mhtech_export_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split the column on the comma and place in a list.
for column in columns:
    column_names = column.split(",")
print(len(column_names))
```

```python
split_data = mhtech_export_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_mhtech_export_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_mhtech_export_df = pd.concat([cleaned_mhtech_export_df, split_data.iloc[:, 4:]], axis=1)
cleaned_mhtech_export_df = cleaned_mhtech_export_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_mhtech_export_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_mhtech_export_df.columns = cleaned_mhtech_export_df.iloc[0]
cleaned_mhtech_export_df.columns = cleaned_mhtech_export_df.columns.str.replace('"', '')  # Remove any quotation marks from column names

# Drop row 0 from the DataFrame
cleaned_mhtech_export_df = cleaned_mhtech_export_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_mhtech_export_df.sample(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_mhtech_export_df = cleaned_mhtech_export_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_mhtech_export_df.head(5)
```

```python
# Convert columns 5 to 68 to float data type
for column in cleaned_mhtech_export_df.columns[5:69]:
    cleaned_mhtech_export_df[column] = pd.to_numeric(cleaned_mhtech_export_df[column], errors='coerce')

# Display the DataFrame with the specified columns converted to float
cleaned_mhtech_export_df.sample(20)
```

```python
cleaned_mhtech_export_df.dtypes
```

```python
# Calculate the average for each row in columns 5 to 68
average = cleaned_mhtech_export_df.iloc[:, 5:69].mean(axis=1)
# Add the average values as a new column named "Average" to the DataFrame
cleaned_mhtech_export_df['MHTexports_%manufac'] = average
cleaned_mhtech_export_df.sample(10)
```
