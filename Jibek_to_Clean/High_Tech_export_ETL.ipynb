```python
import pandas as pd
import os
```

```python
tech_export= "High_Tech_Export.csv"
tech_export_df = pd.read_csv(tech_export, sep="\t")
tech_export_df.head(5)
```

```python
list(tech_export_df)
```

```python
# Get the contents of row 2
row_2_data = tech_export_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split the column on the comma and place in a list.
for column in columns:
    column_names = column.split(",")
print(len(column_names))
```

```python
split_data = tech_export_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_tech_export_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_tech_export_df = pd.concat([cleaned_tech_export_df , split_data.iloc[:, 4:]], axis=1)
cleaned_tech_export_df = cleaned_tech_export_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_tech_export_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_tech_export_df.columns = cleaned_tech_export_df.iloc[0]
cleaned_tech_export_df.columns = cleaned_tech_export_df.columns.str.replace('"', '')  # Remove any quotation marks from column names

# Drop row 0 from the DataFrame
cleaned_tech_export_df = cleaned_tech_export_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_tech_export_df.sample(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_tech_export_df_final = cleaned_tech_export_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_tech_export_df_final.head(5)

```

```python
# # Convert columns 5 to 68 to float data type
# for column in cleaned_tech_export_df_final.columns[5:69]:
#     cleaned_tech_export_df_final[column] = pd.to_numeric(cleaned_tech_export_df_final[column], errors='coerce')

# # Display the DataFrame with the specified columns converted to float
# cleaned_tech_export_df_final.head(5)

# Convert columns 5 to 68 to float data type
for column in cleaned_tech_export_df_final.columns[5:69]:
    cleaned_tech_export_df_final[column] = pd.to_numeric(cleaned_tech_export_df_final[column], errors='coerce')

# Fill non-finite values (NaN) with a default value (e.g., 0)
cleaned_tech_export_df_final.fillna(0, inplace=True)

# Convert the specified columns from float to integer data type
cleaned_tech_export_df_final.iloc[:, 5:69] = cleaned_tech_export_df_final.iloc[:, 5:69].astype(int)

# Display the DataFrame with the specified columns converted to integer
cleaned_tech_export_df_final.head(5)

```

```python
cleaned_tech_export_df_final.dtypes
```

```python
# Calculate the average for each row in columns 5 to 68, skipping NaN values
average = cleaned_tech_export_df_final.iloc[:, 5:69].mean(axis=1, skipna=True)
# Convert the average values to integer data type
average = average.astype(int)
# Add the average values as a new column named "Average_tech_export_$" to the DataFrame
cleaned_tech_export_df_final['Average_tech_export_$'] = average
# Display the DataFrame to verify the changes
cleaned_tech_export_df_final.sample(5)
```

```python
hitech_export_machine_df=cleaned_tech_export_df_final[['Country Name','Average_tech_export_$']]
hitech_export_machine_df.sample(5)
```

```python
# Create the "Output_Jibek" directory if it doesn't exist
output_directory = "../Output_Jibek"
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
# Save the cleaned DataFrame to a CSV file in the specified directory
hitech_export_machine_df.to_csv(os.path.join(output_directory, 'hitech_export_machine_df'), index=False)
```
