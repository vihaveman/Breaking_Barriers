```python
import pandas as pd
import os
```

```python
gdp = "GDP_Growth.csv"
gdp_df = pd.read_csv(gdp, sep="\t")
gdp_df.head(10) 
```

```python
list(gdp_df)
```

```python
# Get the contents of row 2
row_2_data = gdp_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split the column on the comma and place in a list.
for column in columns:
    column_names = column.split(",")
print(len(column_names))
```

```python
split_data = gdp_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_gdp_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_gdp_df = pd.concat([cleaned_gdp_df, split_data.iloc[:, 4:]], axis=1)
cleaned_gdp_df = cleaned_gdp_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_gdp_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_gdp_df.columns = cleaned_gdp_df.iloc[0]
cleaned_gdp_df.columns = cleaned_gdp_df.columns.str.replace('"', '')  # Remove any quotation marks from column names

# Drop row 0 from the DataFrame
cleaned_gdp_df = cleaned_gdp_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_gdp_df.sample(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_gdp_df_final = cleaned_gdp_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_gdp_df_final.head(5)
```

```python
# Convert columns to float data type
for column in cleaned_gdp_df_final.columns[55:66]:
    cleaned_gdp_df_final[column] = pd.to_numeric(cleaned_gdp_df_final[column], errors='coerce')

# Display the DataFrame with the specified columns converted to float
cleaned_gdp_df_final.head(5)
```

```python
cleaned_gdp_df_final.dtypes
```

```python
# Slice the DataFrame from column index 55 to 65 (inclusive)
selected_columns = cleaned_gdp_df_final.iloc[:, 55:66]
# Get the column names of the selected columns
column_names = selected_columns.columns.tolist()
# Print the column names
print(column_names)
```

```python
# Calculate the average for each row in 2015-2022
average = cleaned_gdp_df_final.iloc[:, 55:66].mean(axis=1)
# Add the average values as a new column named "Average" to the DataFrame
cleaned_gdp_df_final['Average_gdp_growth%'] = average
```

```python
cleaned_gdp_df_final
```

```python
gdpgrowth_machine_df=cleaned_gdp_df_final[['Country Name','Average_gdp_growth%']]
gdpgrowth_machine_df.sample(50)
```

```python
# Create the "Output_Jibek" directory if it doesn't exist
output_directory = "../Output_Jibek"
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
# Save the cleaned DataFrame to a CSV file in the specified directory
gdpgrowth_machine_df.to_csv(os.path.join(output_directory, 'gdpgrowth_machine_df'), index=False)
```
