```python
split_data = internet_usage_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_internet_usage_df= pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_internet_usage_df = pd.concat([cleaned_internet_usage_df, split_data.iloc[:, 4:]], axis=1)
cleaned_internet_usage_df = cleaned_internet_usage_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_internet_usage_df.head(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_internet_usage_df_final = cleaned_internet_usage_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_internet_usage_df_final.head(5)
```

```python
# Assign row 0 as column names
cleaned_internet_usage_df.columns = cleaned_internet_usage_df.iloc[0]
cleaned_internet_usage_df.columns = cleaned_internet_usage_df.columns.str.replace('"', '')  # Remove any quotation marks from column names

# Drop row 0 from the DataFrame
cleaned_internet_usage_df = cleaned_internet_usage_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_internet_usage_df.sample(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_internet_usage_df_final = cleaned_internet_usage_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_internet_usage_df_final.head(5)
```

```python
# Convert columns to float data type
for column in cleaned_internet_usage_df_final.columns[60:67]:
    cleaned_internet_usage_df_final[column] = pd.to_numeric(cleaned_internet_usage_df_final[column], errors='coerce')

# Display the DataFrame with the specified columns converted to float
cleaned_internet_usage_df_final.head(5)
```

```python
import numpy as np

# Replace 0 with NaN in the DataFrame
cleaned_internet_usage_df_final.replace(0, np.nan, inplace=True)

# Calculate the latest value for each row in columns , ignoring NaN values and considering the last non-NaN value
if cleaned_internet_usage_df_final.shape[1] >= 67:
    latest = cleaned_internet_usage_df_final.iloc[:, 60:67].apply(lambda row: row.dropna().iloc[-1] if not row.dropna().empty else np.nan, axis=1)
else:
    latest = cleaned_internet_usage_df_final.iloc[:, 60:].apply(lambda row: row.dropna().iloc[-1] if not row.dropna().empty else np.nan, axis=1)

# Add the latest values as a new column named "individuals_%_use_int" to the DataFrame
cleaned_internet_usage_df_final['individuals_%_use_int'] = latest

# Display the DataFrame to verify the changes
cleaned_internet_usage_df_final.sample(5)
```

```python
intusage_machine_df=cleaned_internet_usage_df_final[['Country Name','individuals_%_use_int']]
intusage_machine_df.sample(5)
```
