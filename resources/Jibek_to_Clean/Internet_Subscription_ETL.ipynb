```python
import pandas as pd
import os
```

```python
in_subs = "Internet_Subscription_Per100.csv"
in_subs_df = pd.read_csv(in_subs, sep="\t")
in_subs_df.head(10) 
```

```python
list(in_subs_df)
```

```python
# Get the contents of row 2
row_2_data = in_subs_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split the column on the comma and place in a list.
for column in columns:
    column_names = column.split(",")
print(len(column_names))
```

```python
split_data = in_subs_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_in_subs_df= pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_in_subs_df = pd.concat([cleaned_in_subs_df, split_data.iloc[:, 4:]], axis=1)
cleaned_in_subs_df = cleaned_in_subs_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_in_subs_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_in_subs_df.columns = cleaned_in_subs_df.iloc[0]
cleaned_in_subs_df.columns = cleaned_in_subs_df.columns.str.replace('"', '')  # Remove any quotation marks from column names

# Drop row 0 from the DataFrame
cleaned_in_subs_df = cleaned_in_subs_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_in_subs_df.sample(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_in_subs_df_final = cleaned_in_subs_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_in_subs_df_final.head(5)
```

```python
# Convert columns to float data type
for column in cleaned_in_subs_df_final.columns[55:66]:
    cleaned_in_subs_df_final[column] = pd.to_numeric(cleaned_in_subs_df_final[column], errors='coerce')

# Display the DataFrame with the specified columns converted to float
cleaned_in_subs_df_final.head(5)
```

```python
cleaned_in_subs_df_final.dtypes
```

```python
# Slice the DataFrame from column index 55 to 65 (inclusive)
selected_columns = cleaned_in_subs_df_final.iloc[:, 55:66]
# Get the column names of the selected columns
column_names = selected_columns.columns.tolist()
# Print the column names
print(column_names)
```

```python
import numpy as np

# Replace 0 with NaN in the DataFrame
cleaned_in_subs_df_final.replace(0, np.nan, inplace=True)

# Calculate the latest value for each row in columns, ignoring NaN values and considering the last non-NaN value
if cleaned_in_subs_df_final.shape[1] >= 65:
    latest = cleaned_in_subs_df_final.iloc[:, 55:66].apply(lambda row: row.dropna().iloc[-1] if not row.dropna().empty else np.nan, axis=1)
else:
    latest = cleaned_in_subs_df_final.iloc[:, 55:].apply(lambda row: row.dropna().iloc[-1] if not row.dropna().empty else np.nan, axis=1)

# Add the latest values as a new column named "individuals_%_use_int" to the DataFrame
cleaned_in_subs_df_final['individuals_%_subs_int'] = latest

# Display the DataFrame to verify the changes
cleaned_in_subs_df_final.sample(5)
```

```python
intsubs_machine_df=cleaned_in_subs_df_final[['Country Name','individuals_%_subs_int']]
intsubs_machine_df.sample(50)
```

```python
# Create the "Output_Jibek" directory if it doesn't exist
output_directory = "../Output_Jibek"
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
# Save the cleaned DataFrame to a CSV file in the specified directory
intsubs_machine_df.to_csv(os.path.join(output_directory, 'intsubs_machine_df'), index=False)
```
