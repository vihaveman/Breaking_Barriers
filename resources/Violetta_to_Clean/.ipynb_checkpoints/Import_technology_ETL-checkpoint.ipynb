```python
import pandas as pd
import os
```

```python
import_tech = "Import_technology.csv"
import_tech_df = pd.read_csv(import_tech, sep="\t")
import_tech_df
```

```python
list(import_tech_df)
```

```python
# Get the contents of row 2
row_2_data = import_tech_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
split_data = import_tech_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_import_tech_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_import_tech_df = pd.concat([cleaned_import_tech_df, split_data.iloc[:, 4:]], axis=1)
cleaned_import_tech_df = cleaned_import_tech_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_import_tech_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_import_tech_df.columns = cleaned_import_tech_df.iloc[0]

# Remove any quotation marks from column names
cleaned_import_tech_df.columns = cleaned_import_tech_df.columns.str.replace('"', '')  

# Drop row 0 from the DataFrame
cleaned_import_tech_df = cleaned_import_tech_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_import_tech_df.head(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_import_tech_df = cleaned_import_tech_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_import_tech_df.head(5)
```

```python
#Filtered columns for desired years
years_columns = [str(year) for year in range(2000, 2023)]

# Select the desired columns
selected_columns = ['Country Name', 'Country Code'] + years_columns

# Create the new DataFrame
selected_columns = cleaned_import_tech_df[selected_columns]

cleaned_import_tech_df = selected_columns

cleaned_import_tech_df.head(10)
```

```python
# Clean DF and Convert Columns

# Remove the quotation marks from all columns
cleaned_import_tech_df = cleaned_import_tech_df.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# Convert columns the Years to float data type
for column in cleaned_import_tech_df.columns[2:25]:
    cleaned_import_tech_df[column] = pd.to_numeric(cleaned_import_tech_df[column], errors='coerce')
    
# Display the updated DataFrame
cleaned_import_tech_df.head()
```

```python
# Step 3: Perform data interpolation on the numeric columns
cleaned_import_tech_df = cleaned_import_tech_df.interpolate()
cleaned_import_tech_df.head(5)
```

```python
# Step 1: Melt the DataFrame to transform the years into a separate column
melted_df = cleaned_import_tech_df.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="High Tech Import")

# Step 2: Sort the DataFrame by "Country Name" and "Year" for better organization
melted_df = melted_df.sort_values(by=["Year", "Country Name"])

# Step 3: Reset the index of the DataFrame
melted_df = melted_df.reset_index(drop=True)

# Round all numerical values to two decimal places
melted_df = melted_df.round(2)

# Step 4: Display the modified DataFrame
melted_df.head()
```

```python
# Export the DataFrame to the CSV file
melted_df.to_csv('Cleaned_Data/High_Tech_Import.csv', index=False)
```
