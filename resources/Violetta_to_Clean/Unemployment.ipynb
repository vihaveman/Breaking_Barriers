```python
#Import Dependancies
import pandas as pd
import os
```

```python
#Import File
unemployment = 'Unemployment.csv'
unemployment_df = pd.read_csv(unemployment,  sep="\t")
unemployment_df.head(10)                                 
```

```python
list(unemployment_df)
```

```python
# Get the contents of row 2
row_2_data = unemployment_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split columns
split_data = unemployment_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
split_unemployment_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)


# Concatenate the remaining columns from original data
split_unemployment_df = pd.concat([split_unemployment_df, split_data.iloc[:, 4:]], axis=1)
split_unemployment_df = split_unemployment_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
split_unemployment_df.head(10)
```

```python
# Assign row 0 as column names
split_unemployment_df.columns = split_unemployment_df.iloc[0]

# Remove any quotation marks from column names
split_unemployment_df.columns = split_unemployment_df.columns.str.replace('"', '')

# Drop row 0 from the DataFrame
split_unemployment_df = split_unemployment_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
split_unemployment_df.head(10)
```

```python
#Filtered columns for desired years
years_columns = [str(year) for year in range(2010, 2021)]

# Select the desired columns
selected_columns = ['Country Name', 'Country Code'] + years_columns

# Create the new DataFrame
selected_columns = split_unemployment_df[selected_columns]

cleaned_unemployment_df = selected_columns

cleaned_unemployment_df.head(10)
```

```python
# Clean DF and Convert Columns

# Remove the quotation marks from all columns
cleaned_unemployment_df = cleaned_unemployment_df.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# Convert columns the Years to float data type
for column in cleaned_unemployment_df.columns[2:13]:
    cleaned_unemployment_df[column] = pd.to_numeric(cleaned_unemployment_df[column], errors='coerce')

# Display the updated DataFrame
cleaned_unemployment_df.head()
```

```python
# Replace the NaN values with 0
cleaned_unemployment_df.fillna(0, inplace=True)

cleaned_unemployment_df.head(10)
```

```python
# Calculate the average for each country
average_by_country = cleaned_unemployment_df.groupby('Country Code').apply(lambda group: group.drop(columns=['Country Name', 'Country Code']).mean(skipna=True, axis=1))

# Replace all 0.0 values with the corresponding country average
for col in cleaned_unemployment_df.columns[2:]:
    cleaned_unemployment_df[col] = cleaned_unemployment_df.apply(lambda row: average_by_country[row['Country Code']][row.name] if row[col] == 0.0 else row[col], axis=1)

cleaned_unemployment_df.head(10)
```

```python
# Step 1: Melt the DataFrame to transform the years into a separate column
melted_df = cleaned_unemployment_df.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="Unemployment Rate")

# Step 2: Sort the DataFrame by "Country Name" and "Year" for better organization
melted_df = melted_df.sort_values(by=["Year","Country Name"])

# Step 3: Reset the index of the DataFrame
melted_df = melted_df.reset_index(drop=True)

# Round all numerical values to two decimal places
melted_df = melted_df.round(2)

# Step 4: Display the modified DataFrame
melted_df.head()
```

```python
# Export the DataFrame to the CSV file
melted_df.to_csv('Cleaned_Data/Unemployment.csv', index=False)
```
