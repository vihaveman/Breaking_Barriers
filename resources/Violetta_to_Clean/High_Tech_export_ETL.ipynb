```python
import pandas as pd
import os
```

```python
tech_export= "High_Tech_Export.csv"
tech_export_df = pd.read_csv(tech_export, sep="\t")
tech_export_df.head(5)
```

```python
list(tech_export_df)
```

```python
# Get the contents of row 2
row_2_data = tech_export_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
split_data = tech_export_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_tech_export_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_tech_export_df = pd.concat([cleaned_tech_export_df , split_data.iloc[:, 4:]], axis=1)
cleaned_tech_export_df = cleaned_tech_export_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_tech_export_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_tech_export_df.columns = cleaned_tech_export_df.iloc[0]

# Remove any quotation marks from column names
cleaned_tech_export_df.columns = cleaned_tech_export_df.columns.str.replace('"', '')  

# Drop row 0 from the DataFrame
cleaned_tech_export_df = cleaned_tech_export_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_tech_export_df.head(5)
```

```python
#Filtered columns for desired years
years_columns = [str(year) for year in range(2000, 2023)]

# Select the desired columns
selected_columns = ['Country Name', 'Country Code'] + years_columns

# Create the new DataFrame
selected_columns = cleaned_tech_export_df[selected_columns]

cleaned_tech_export_df = selected_columns

cleaned_tech_export_df.head(10)
```

```python
# Clean DF and Convert Columns

# Remove the quotation marks from all columns
cleaned_tech_export_df = cleaned_tech_export_df.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# Convert columns the Years to float data type
for column in cleaned_tech_export_df.columns[2:13]:
    cleaned_tech_export_df[column] = pd.to_numeric(cleaned_tech_export_df[column], errors='coerce')
    
## Generate the list of columns from "2000" to "2022"
start_year = 2000
end_year = 2022
numerical_columns = [str(year) for year in range(start_year, end_year + 1)]

#Convert the columns to numeric values
cleaned_tech_export_df[numerical_columns] = cleaned_tech_export_df[numerical_columns].apply(pd.to_numeric)

# Divide the columns by 1 billion (1000000)
cleaned_tech_export_df[numerical_columns] = cleaned_tech_export_df[numerical_columns].div(1000000)

# Display the updated DataFrame
cleaned_tech_export_df.head()
```

```python
# Step 3: Perform data interpolation on the numeric columns
cleaned_tech_export_df = cleaned_tech_export_df.interpolate()
cleaned_tech_export_df.head(5)
```

```python
# Step 1: Melt the DataFrame to transform the years into a separate column
melted_df = cleaned_tech_export_df.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="High Tech Export ($M)")

# Step 2: Sort the DataFrame by "Country Name" and "Year" for better organization
melted_df = melted_df.sort_values(by=["Year", "Country Name"])

# Step 3: Reset the index of the DataFrame
melted_df = melted_df.reset_index(drop=True)

# Round all numerical values to two decimal places
melted_df = melted_df.round(2)

# Step 4: Display the modified DataFrame
melted_df.head()
```

```python
# Export the DataFrame to the CSV file
melted_df.to_csv('Cleaned_Data/High_Tech_Export.csv', index=False)
```
