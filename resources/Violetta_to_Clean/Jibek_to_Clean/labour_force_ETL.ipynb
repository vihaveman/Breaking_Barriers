```python
import pandas as pd
import os
```

```python
labforce = "Labour_Force.csv"
labforce_df = pd.read_csv(labforce, sep="\t")
labforce_df
```

```python
# Get the contents of row 2
row_2_data = labforce_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split the column on the comma and place in a list.
for column in columns:
    column_names = column.split(",")
print(len(column_names))
```

```python
split_data = labforce_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_labforce_df= pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_labforce_df = pd.concat([cleaned_labforce_df, split_data.iloc[:, 4:]], axis=1)
cleaned_labforce_df = cleaned_labforce_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_labforce_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_labforce_df.columns = cleaned_labforce_df.iloc[0]
cleaned_labforce_df.columns = cleaned_labforce_df.columns.str.replace('"', '')  # Remove any quotation marks from column names

# Drop row 0 from the DataFrame
cleaned_labforce_df = cleaned_labforce_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_labforce_df.sample(5)
```

```python
# Define a function to remove double quotes from each cell
def remove_quotes(cell_value):
    if cell_value is not None:
        return cell_value.replace('"', '')
    else:
        return cell_value

# Use applymap() to remove double quotes from the entire DataFrame
cleaned_labforce_df_final = cleaned_labforce_df.applymap(remove_quotes)

# Display the first five rows of the cleaned DataFrame
cleaned_labforce_df_final.head(5)
```

```python
# Convert columns to float data type
for column in cleaned_labforce_df_final.columns[60:67]:
    cleaned_labforce_df_final[column] = pd.to_numeric(cleaned_labforce_df_final[column], errors='coerce')

# Fill non-finite values (NaN) with a default value (e.g., 0)
cleaned_labforce_df_final.fillna(0, inplace=True)

# Convert the specified columns from float to integer data type
cleaned_labforce_df_final.iloc[:, 60:67] = cleaned_labforce_df_final.iloc[:, 60:67].astype(int)

# Display the DataFrame with the specified columns converted to integer
cleaned_labforce_df_final.head(5)
```

```python
# Calculate the average for each row in columns 5 to 68, skipping NaN values
average = cleaned_labforce_df_final.iloc[:, 60:67].mean(axis=1, skipna=True)
# Convert the average values to integer data type
average = average.astype(int)
# Add the average values as a new column named "Average_tech_export_$" to the DataFrame
cleaned_labforce_df_final['lab_for_pop'] = average
# Display the DataFrame to verify the changes
cleaned_labforce_df_final.sample(5)
```
