```python
#Import Dependancies
import pandas as pd
import os
```

```python
#Import File
mobile_sub = 'Mobile_Subscription_per100.csv'
mobile_sub_df = pd.read_csv(mobile_sub,  sep="\t")
mobile_sub_df.head(10)                                 
```

```python
list(mobile_sub_df)
```

```python
# Get the contents of row 2
row_2_data = mobile_sub_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split columns
split_data = mobile_sub_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
split_mobile_sub_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)


# Concatenate the remaining columns from original data
split_mobile_sub_df = pd.concat([split_mobile_sub_df, split_data.iloc[:, 4:]], axis=1)
split_mobile_sub_df = split_mobile_sub_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
split_mobile_sub_df.head(10)
```

```python
# Assign row 0 as column names
split_mobile_sub_df.columns = split_mobile_sub_df.iloc[0]

# Remove any quotation marks from column names
split_mobile_sub_df.columns = split_mobile_sub_df.columns.str.replace('"', '')

# Drop row 0 from the DataFrame
split_mobile_sub_df = split_mobile_sub_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
split_mobile_sub_df.head(10)
```

```python
#Filtered columns for desired years
years_columns = [str(year) for year in range(2000, 2023)]

# Select the desired columns
selected_columns = ['Country Name', 'Country Code'] + years_columns

# Create the new DataFrame
selected_columns = split_mobile_sub_df[selected_columns]

cleaned_mobile_sub_df = selected_columns

cleaned_mobile_sub_df.head(10)
```

```python
# Clean DF and Convert Columns

# Remove the quotation marks from all columns
cleaned_mobile_sub_df = cleaned_mobile_sub_df.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# Convert columns the Years to float data type
for column in cleaned_mobile_sub_df.columns[2:25]:
    cleaned_mobile_sub_df[column] = pd.to_numeric(cleaned_mobile_sub_df[column], errors='coerce')

# Display the updated DataFrame
cleaned_mobile_sub_df.head()
```

```python
# Perform data interpolation on the numeric columns
cleaned_mobile_sub_df = cleaned_mobile_sub_df.interpolate()
cleaned_mobile_sub_df.head(5)
```

```python
# Step 1: Melt the DataFrame to transform the years into a separate column
melted_df = cleaned_mobile_sub_df.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="Mobile Sub Per 100")

# Step 2: Sort the DataFrame by "Country Name" and "Year" for better organization
melted_df = melted_df.sort_values(by=["Year", "Country Name"])

# Step 3: Reset the index of the DataFrame
melted_df = melted_df.reset_index(drop=True)

# Round all numerical values to two decimal places
melted_df = melted_df.round(2)

# Step 4: Display the modified DataFrame
melted_df.head()
```

```python
# Export the DataFrame to the CSV file
melted_df.to_csv('Cleaned_Data/Mobile Subscription.csv', index=False)
```
