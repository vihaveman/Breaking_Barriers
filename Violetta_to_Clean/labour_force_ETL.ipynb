```python
import pandas as pd
import os
```

```python
labforce = "Labour_Force.csv"
labforce_df = pd.read_csv(labforce, sep="\t")
labforce_df
```

```python
list(labforce_df)
```

```python
# Get the contents of row 2
row_2_data = labforce_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
split_data = labforce_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
cleaned_labforce_df= pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)

# Concatenate the remaining columns from split_data to cleaned_atm_df
cleaned_labforce_df = pd.concat([cleaned_labforce_df, split_data.iloc[:, 4:]], axis=1)
cleaned_labforce_df = cleaned_labforce_df.drop(index=0).reset_index(drop=True)
# Display the cleaned DataFrame
cleaned_labforce_df.head(5)
```

```python
# Assign row 0 as column names
cleaned_labforce_df.columns = cleaned_labforce_df.iloc[0]

# Remove any quotation marks from column names
cleaned_labforce_df.columns = cleaned_labforce_df.columns.str.replace('"', '')  

# Drop row 0 from the DataFrame
cleaned_labforce_df = cleaned_labforce_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
cleaned_labforce_df.sample(5)
```

```python
#Filtered columns for desired years
years_columns = [str(year) for year in range(2010, 2021)]

# Select the desired columns
selected_columns = ['Country Name', 'Country Code'] + years_columns

# Create the new DataFrame
selected_columns = cleaned_labforce_df[selected_columns]

cleaned_labforce_df = selected_columns

cleaned_labforce_df.head(10)
```

```python
# Clean DF and Convert Columns

# Remove the quotation marks from all columns
cleaned_labforce_df = cleaned_labforce_df.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# Convert columns the Years to float data type
for column in cleaned_labforce_df.columns[2:13]:
    cleaned_labforce_df[column] = pd.to_numeric(cleaned_labforce_df[column], errors='coerce')
    
# List of columns from "2010" to "2020"
numerical_columns = ["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]

# Divide the columns by 1 billion
cleaned_labforce_df[numerical_columns] = cleaned_labforce_df[numerical_columns].div(1000000)

# Display the updated DataFrame
cleaned_labforce_df.head()
```

```python
# Replace the NaN values with 0
cleaned_labforce_df.fillna(0, inplace=True)

cleaned_labforce_df.head(10)
```

```python
# Calculate the average for each country
average_by_country = cleaned_labforce_df.groupby('Country Code').apply(lambda group: group.drop(columns=['Country Name', 'Country Code']).mean(skipna=True, axis=1))

# Replace all 0.0 values with the corresponding country average
for col in cleaned_labforce_df.columns[2:]:
    cleaned_labforce_df[col] = cleaned_labforce_df.apply(lambda row: average_by_country[row['Country Code']][row.name] if row[col] == 0.0 else row[col], axis=1)

cleaned_labforce_df.head(10)
```

```python
# Step 1: Melt the DataFrame to transform the years into a separate column
melted_df = cleaned_labforce_df.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="Labour Force (M)")

# Step 2: Sort the DataFrame by "Country Name" and "Year" for better organization
melted_df = melted_df.sort_values(by=["Year", "Country Name"])

# Step 3: Reset the index of the DataFrame
melted_df = melted_df.reset_index(drop=True)

# Round all numerical values to two decimal places
melted_df = melted_df.round(2)

# Step 4: Display the modified DataFrame
melted_df.head()
```

```python
# Export the DataFrame to the CSV file
melted_df.to_csv('Cleaned_Data/Labour_Force.csv', index=False)
```
