```python
#Import Dependancies
import pandas as pd
import os
```

```python
#Import File
ip_income = 'Recipient_of IP_Income.csv'
ip_income_df = pd.read_csv(ip_income,  sep="\t")
ip_income_df.head(10)                                 
```

```python
list(ip_income_df)
```

```python
# Get the contents of row 2
row_2_data = ip_income_df.iloc[1]
# Convert row 2 data to a list
columns = row_2_data.tolist()
columns
```

```python
# Split columns
split_data = ip_income_df['Data Source,"World Development Indicators",'].str.split(',', n=68, expand=True)

# Create a new DataFrame to store the cleaned data with appropriate column names
split_ip_income_df = pd.DataFrame(
    {
        'Country Name': split_data[0],
        'Country Code': split_data[1],
        'Indicator Name': split_data[2],
        'Indicator Code': split_data[3],
    }
)


# Concatenate the remaining columns from original data
split_ip_income_df = pd.concat([split_ip_income_df, split_data.iloc[:, 4:]], axis=1)
split_ip_income_df = split_ip_income_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
split_ip_income_df.head(10)
```

```python
# Assign row 0 as column names
split_ip_income_df.columns = split_ip_income_df.iloc[0]

# Remove any quotation marks from column names
split_ip_income_df.columns = split_ip_income_df.columns.str.replace('"', '')

# Drop row 0 from the DataFrame
split_ip_income_df = split_ip_income_df.drop(index=0).reset_index(drop=True)

# Display the cleaned DataFrame
split_ip_income_df.head(10)
```

```python
#Filtered columns for desired years
years_columns = [str(year) for year in range(2010, 2021)]

# Select the desired columns
selected_columns = ['Country Name', 'Country Code'] + years_columns

# Create the new DataFrame
selected_columns = split_ip_income_df[selected_columns]

cleaned_ip_income_df = selected_columns

cleaned_ip_income_df.head(10)
```

```python
# Clean DF and Convert Columns

# Remove the quotation marks from all columns
cleaned_ip_income_df = cleaned_ip_income_df.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# Convert columns the Years to float data type
for column in cleaned_ip_income_df.columns[2:13]:
    cleaned_ip_income_df[column] = pd.to_numeric(cleaned_ip_income_df[column], errors='coerce')
    
# List of columns from "2010" to "2020"
numerical_columns = ["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]    
    
# Divide the columns by 1 billion
cleaned_ip_income_df[numerical_columns] = cleaned_ip_income_df[numerical_columns].div(1000000)    

# Round all numerical values to two decimal places
cleaned_ip_income_df = cleaned_ip_income_df.round(2)

# Display the updated DataFrame
cleaned_ip_income_df.head()

```

```python
# Step 1: Melt the DataFrame to transform the years into a separate column
melted_df = cleaned_ip_income_df.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="IP Income($M)")

# Step 2: Sort the DataFrame by "Country Name" and "Year" for better organization
melted_df = melted_df.sort_values(by=["Year", "Country Name"])

# Step 3: Reset the index of the DataFrame
melted_df = melted_df.reset_index(drop=True)

# Step 4: Display the modified DataFrame
melted_df.head()
```

```python
# Export the DataFrame to the CSV file
melted_df.to_csv('Cleaned_Data/IP_Income.csv', index=False)
```
