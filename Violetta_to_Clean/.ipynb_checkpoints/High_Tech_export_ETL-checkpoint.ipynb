```python
# Clean DF and Convert Columns

# Remove the quotation marks from all columns
cleaned_tech_export_df_final = cleaned_tech_export_df_final.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# Convert columns the Years to float data type
for column in cleaned_tech_export_df_final.columns[2:13]:
    cleaned_tech_export_df_final[column] = pd.to_numeric(cleaned_tech_export_df_final[column], errors='coerce')   
    
# Display the updated DataFrame
cleaned_tech_export_df_final.head()
```

```python
# Replace the NaN values with 0
cleaned_tech_export_df_final.fillna(0, inplace=True)

cleaned_tech_export_df_final.head(10)
```

```python
# Calculate the average for each country
average_by_country = cleaned_tech_export_df_final.groupby('Country Code').apply(lambda group: group.drop(columns=['Country Name', 'Country Code']).mean(skipna=True, axis=1))

# Replace all 0.0 values with the corresponding country average
for col in cleaned_tech_export_df_final.columns[2:]:
    cleaned_tech_export_df_final[col] = cleaned_tech_export_df_final.apply(lambda row: average_by_country[row['Country Code']][row.name] if row[col] == 0.0 else row[col], axis=1)

cleaned_tech_export_df_final.head(10)
```

```python
# Step 1: Melt the DataFrame to transform the years into a separate column
melted_df = cleaned_tech_export_df_final.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="High Tech Export")

# Step 2: Sort the DataFrame by "Country Name" and "Year" for better organization
melted_df = melted_df.sort_values(by=["Year", "Country Name"])

# Step 3: Reset the index of the DataFrame
melted_df = melted_df.reset_index(drop=True)

# Round all numerical values to two decimal places
melted_df = melted_df.round(2)

# Step 4: Display the modified DataFrame
melted_df.head()
```

```python
# Export the DataFrame to the CSV file
melted_df.to_csv('Cleaned_Data/High_Tech_Export.csv', index=False)
```
