```python
import pandas as pd
```

```python
# Combining 2018 files 
merged_data = ('Cleaned_Data/Cleaned_Mobile_Sub.csv', 'Cleaned_Data/Med_High_Tech_Manufacturing.csv',
             'Cleaned_Data/Mobile_Subcrition_Per100.csv', 'Cleaned_Data/Personal_Remittance.csv')

# Creating an empty list to store dataframe
merged_list = []

#Looping through files and store into data frame
for file in merged_data:
    merged_data = pd.read_csv(file)
    merged_list.append(merged_data)

merged_df = pd.concat(merged_list)

merged_df
```

```python
# Export the DataFrame to the CSV file
merged_df.to_csv('Cleaned_Data/Merged_Data.csv', index=False)
```
